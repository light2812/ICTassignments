We will drop 'No of Times Visited' cuz of missing values

```python
df.drop('No of Times Visited',axis=1,inplace=True)
```

```python
miss_cols=df.columns[df.isna().any()]
miss_cols
```

```python
for col in miss_cols:
  plt.figure()
  sns.histplot(df[col],kde=True)
```

Now we fill missing values in each

```python
df['Sale Price'].fillna(df['Sale Price'].median(),inplace=True)
```

```python

df['Flat Area (in Sqft)'].fillna(df['Flat Area (in Sqft)'].median(),inplace=True)
df['Lot Area (in Sqft)'].fillna(df['Lot Area (in Sqft)'].median(),inplace=True)
df['No of Bathrooms'].fillna(df['No of Bathrooms'].mode()[0],inplace=True)
```

```python
df['Area of the House from Basement (in Sqft)'].fillna(df['Area of the House from Basement (in Sqft)'].median(),inplace=True)

df['Zipcode'].fillna(df['Zipcode'].mode()[0],inplace=True)
df['Latitude'].fillna(df['Latitude'].mean(),inplace=True)
df['Longitude'].fillna(df['Longitude'].mean(),inplace=True)
df['Living Area after Renovation (in Sqft)'].fillna(df['Living Area after Renovation (in Sqft)'].median(),inplace=True)
```

```python
df.isna().sum()
```

MISSING VALUES HAVE BEEN FILLED

```python
df.head()
```

OUTLIER DETECTION


We use a loop to draw box plots for all numeric columns

```python
for col in df.columns:
  if df[col].dtype!='object':
    plt.figure()
    sns.boxplot(df[col])
```

now we remove outliers from all of them

```python
for col in df.columns:
  if df[col].dtype!='object':
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    upper_limit=q3+1.5*iqr
    lower_limit=q1-1.5*iqr
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)
    plt.figure()
    sns.boxplot(df[col])
```

OUTLIERS HAVE BEEN REMOVED
