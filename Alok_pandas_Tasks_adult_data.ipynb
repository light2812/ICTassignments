# **Create a new column 'income_binary' where <=50K is 0 and >50K is 1.**

```python
d1['income_binary']=''
d1.loc[d1['income']=='<=50K','income_binary']=0
d1.loc[d1['income']=='>50K','income_binary']=1
d1.head()
```

# **Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).**

```python
d1['age_group']=''
d1.loc[d1['age']<30,'age_group']='Youth'
d1.loc[(d1['age']>=30) & (d1['age']<60) ,'age_group']='Adult'
d1.loc[d1['age']>=60,'age_group']='Senior'
d1.head()
```

# **Rename the 'educational-num' column to 'education_num'.**

```python
d1 = d1.rename(columns={'educational-num': 'education_num'})
d1.head()
```

# **Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.**

```python
d1['workclass'].replace('?', 'Unknown', inplace=True)
d1['occupation'].replace('?', 'Unknown', inplace=True)
```
