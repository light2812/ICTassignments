Q9:: Based on your analysis, provide a brief summary of any insights or patterns
you discovered in the dataset.

```python
genre_profitability = df.groupby('Genre')['Profitability'].mean().sort_values(ascending=False)
genre_profitability.plot(kind='bar', color='skyblue')
plt.title('Average Profitability by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Profitability')
plt.xticks(rotation=45)
plt.show()
```

The analysis reveals which genres tend to have higher profitability, offering insights into audience preferences and market trends.
In this case, Horror is most profitable and Mystery is the least profitable.
